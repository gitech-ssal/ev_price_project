```python
# ----------------------------------------------------------------------------------------------------
# 라이브러리 목록

# 기본 라이브러리 
import pandas as pd

# 시각화 라이브러리 
import matplotlib
import matplotlib.pyplot as plt

# 개인 라이브러리 
from preparation_for_analysis.show_window import DataVisualizer

# pandas 설정
pd.set_option('display.max_rows', None)  # 모든 행 출력
pd.set_option('display.max_columns', None)  # 모든 열 출력

# matplotlib 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # 윈도우에서 사용하는 한글 폰트
matplotlib.rcParams['axes.unicode_minus'] = False  # 음수 기호 문제 해결

# preparation_for_analysis 설정
visualizer = DataVisualizer(line="=", length=100, start="#")
# ----------------------------------------------------------------------------------------------------
```

```python
# ----------------------------------------------------------------------------------------------------
# rmse 불러오기 
rmse = pd.read_csv("C:\\Users\\ssalt\\Documents\\ev_price_predict_project\\data\\train\\A_df\\e_model\\model_1\\2_rmse.csv",
                    sep=",",
                    header=0)

visualizer.show_df_info(title="rmse.info", df=rmse)
# ----------------------------------------------------------------------------------------------------
```

```python
# ----------------------------------------------------------------------------------------------------
# rmse 시각화
# line plot으로 시각화 

# 데이터 준비 
epochs = rmse["Epoch"]
values = rmse["RMSE"]
epochs_labels = rmse["Epoch"].apply(lambda x: "Epoch "+str(x))

# 그래프 생성
plt.figure(figsize=(30, 6))

# 선 그래프 그리기
plt.plot(epochs, values, marker='o', color='#C60C30', linestyle='-', linewidth=2, label='RMSE')

# 그래프 제목과 축 레이블 설정
plt.title("Epoch vs RMSE", fontsize=16, fontweight='bold')
plt.xlabel("Epoch", fontsize=14)
plt.ylabel("RMSE", fontsize=14)

# Y축 범위 설정
plt.ylim(0, 5)

# 범례 추가
plt.legend(fontsize=12)

# 격자 추가
plt.grid(linestyle='--', alpha=0.6)

# 그래프 출력
plt.tight_layout()
plt.show()
```
