```python
# ----------------------------------------------------------------------------------------------------
# 라이브러리 목록
# 기본 라이브러리 
import pandas as pd

# sklearn 라이브러리
from sklearn.model_selection import train_test_split

# pandas 설정
pd.set_option('display.max_rows', None)  # 모든 행 출력
pd.set_option('display.max_columns', None)  # 모든 열 출력

# 개인 라이브러리  
from preparation_for_analysis.show_window import DataVisualizer

# preparation_for_analysis 설정
visualizer = DataVisualizer(line="=", length=100, start="#")
# ----------------------------------------------------------------------------------------------------
```

```python
# ----------------------------------------------------------------------------------------------------
# revision 불러오기 
revision = pd.read_csv("C:\\Users\\ssalt\\Documents\\ev_price_predict_project\\data\\train\\A_df\\b_revision\\1_revision.csv",
                       sep=",",
                       header=0)

visualizer.show_df_info(title="revision", df=revision)
# ----------------------------------------------------------------------------------------------------
```

```python
# ----------------------------------------------------------------------------------------------------
# revision을 learning과 validation으로 분리 
learning, validation = train_test_split(revision, test_size=0.1, random_state=1234)

# learning 재설정 
learning.reset_index(drop=True, inplace=True)

# validation 재설정
validation.reset_index(drop=True, inplace=True)

# learning 정보 확인 
visualizer.show_df_info(title="learning", df=learning)

# validation 정보 확인 
visualizer.show_df_info(title="validation", df=validation)
# ----------------------------------------------------------------------------------------------------
```

```python
# ----------------------------------------------------------------------------------------------------
# learning 저장 
learning.to_csv("C:\\Users\\ssalt\\Documents\\ev_price_predict_project\\data\\train\\A_df\\c_learning\\1_learning.csv",
                index=False)

# validation 저장 
validation.to_csv("C:\\Users\\ssalt\\Documents\\ev_price_predict_project\\data\\train\\A_df\\d_validation\\1_validation.csv",
                  index=False)
# ----------------------------------------------------------------------------------------------------
```
